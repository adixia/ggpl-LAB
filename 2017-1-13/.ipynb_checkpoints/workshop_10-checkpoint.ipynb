```python
import copy
"""DA LAVORO _ poi DA trasferire in my function"""
def generate_beam_structure(dir_name,floor,hasStair=0):
    result=generate_sub_beam("./"+dir_name+"/beam_pillar.lines")
    beams=(result[1])
    beams=PROD([beams,QUOTE([2])])
    pillar=(generate_pillar(result[0]))
    myFloor=floor
    
#toglie pavimento per le scale
    if(hasStair!=0):
        stairCuboid=generate_sub_beam("./lines/stairs.lines")
        stairsPos=stairCuboid[0]
        stairCuboid=SOLIDIFY(stairCuboid[1])
        size=SIZE([1,2])(stairCuboid)
        print "la size è ",size
        stairCuboid=CUBOID([size[0]/2.0,size[1]])
        if(hasStair==3):
            stairCuboid=CUBOID([size[0],size[1]])
            stairCuboid=STRUCT([T([1,2])([stairsPos[0][0],stairsPos[0][1]]),stairCuboid])
        if(hasStair==2):
            stairCuboid=STRUCT([T([1,2])([stairsPos[0][0]+size[0]/2.0,stairsPos[0][1]]),stairCuboid])
            print "quando 2",SIZE([1,2])(stairCuboid) 
        if(hasStair==1):
            stairCuboid=STRUCT([T([1,2])([stairsPos[0][0],stairsPos[0][1]]),stairCuboid])
            print "quando 1 ",SIZE([1,2])(stairCuboid)
        myFloor=DIFF([myFloor,stairCuboid])
    beams_and_pillar=STRUCT([pillar,T(3)(SIZE(3)(pillar)),beams,T(3)(SIZE(3)(beams)),(myFloor)])
    beams_and_pillar=(OFFSET([1,1,2])(beams_and_pillar))
    return beams_and_pillar


def findStair():
    result=generate_sub_beam("./lines/stairs.lines")
    return result[1]
    

def generate_pillar(points):
    newP=[]
    facetes=[]
    i=1
    for p in points:
        p1=copy.copy(p)
        p1.append(0)
        p2=copy.copy(p)
        p2.append(34)
        facetes.append([i,i+1])
        newP.append(p1)
        newP.append(p2)
        i=i+2
   
    return (MKPOL([newP,facetes,None]))
        
    
    
def generate_sub_beam(file_name):
    i=0
    points=[]
    indesxs=[]
    pillar=[]
    
    with open(file_name, 'rb') as csvfile:
        spamreader = csv.reader(csvfile)
        for row in spamreader:
            points.append([float(row[0]),float(row[1])])
            points.append([float(row[2]),float(row[3])])
            i=i+2
            indesxs.append([i-1,i])
    
    
    result=[points,MKPOL([points,indesxs,None])]
    return result
```
